```python
from sqlalchemy import create_engine
from sklearn import tree, metrics
from IPython.display import Image
from subprocess import check_call
from pandas import *
import numpy
import matplotlib.pylab as plot

%matplotlib inline
```

```python
# Connect to Postgres
rdbms = 'postgresql'
host = 'localhost'
port = '5432'
database = 'jupyter'
user = 'jupyter'
password = 'jupyter'

credentials = (rdbms, user, password, host, port, database)
connection_string = "%s://%s:%s@%s:%s/%s" % credentials
engine = create_engine(connection_string)
```

```python
# Load data from Postgres to dataframe
query = 'select * from iris'
iris_data = pandas.read_sql_query(query, con=engine)
```

```python
# Create dummy variables for categories
code = {
    'Iris-setosa': 0,
    'Iris-virginica': 1,
    'Iris-versicolor': 2
}
iris_data['code'] = iris_data['class'].apply(lambda x: code[x])
```

```python
# Select training, test data
training = iris_data.sample(120)
test = iris_data.drop(training.index.values)
```

```python
# Massage training data
dimensions = training.columns.values[:4]
training_data = training.as_matrix(dimensions)
training_target = training['class'].values
```

```python
# Massage test data
dimensions = test.columns.values[:4]
test_data = test.as_matrix(dimensions)
test_target = test['class'].values
```

```python
# Train decision tree
classifier = tree.DecisionTreeClassifier(max_depth=3, criterion='entropy')
classifier = classifier.fit(training_data, training_target)
```

```python
# Make predictions
predicted = classifier.predict(test_data)
expected = test_target
```

```python
# Evaluate results
print(metrics.classification_report(expected, predicted))
print("\nConfusion Matrix")
print(metrics.confusion_matrix(expected, predicted))
```

```python
# Visualize decision tree
# Requires graphviz ($ brew install graphviz)
tree.export_graphviz(
    classifier,
    out_file='tree.dot',
    feature_names = iris_data.columns.values[:4],
    class_names=unique(iris_data['class'].values),
    filled=True,
    rounded=True,
    special_characters=True
)

command = ['dot', '-Tpng', 'tree.dot', '-o', 'tree.png']
check_call(command)
Image('tree.png')
```

```python
# Scatter plot in blossom space
species = iris_data.groupby('class')
fig, ax = plot.subplots()

x, y = 'petal_length', 'petal_width'

flat_data = training.as_matrix([x, y])
flat_target = training['code']

flat_classifier = tree.DecisionTreeClassifier(
    max_depth=4,
    criterion='entropy'
).fit(flat_data, flat_target)

plot_step = 0.02
x_min, x_max = flat_data[:, 0].min() - 1, flat_data[:, 0].max() + 1
y_min, y_max = flat_data[:, 1].min() - 1, flat_data[:, 1].max() + 1
xx, yy = numpy.meshgrid(
    numpy.arange(x_min, x_max, plot_step),
    numpy.arange(y_min, y_max, plot_step)
)

Z = flat_classifier.predict(numpy.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
cs = plot.contourf(xx, yy, Z, cmap=plot.cm.Paired)

for name, group in species:
    ax.plot(
        group[x],
        group[y],
        linestyle='',
        marker='o',
        label=name
    )
```

```python
# Visualize decision tree
# Requires graphviz ($ brew install graphviz)
tree.export_graphviz(
    flat_classifier,
    out_file='flat-tree.dot',
    feature_names = [x, y],
    class_names=unique(training['class'].values),
    filled=True
)

command = ['dot', '-Tpng', 'flat-tree.dot', '-o', 'flat-tree.png']
check_call(command)
Image('flat-tree.png')
```
